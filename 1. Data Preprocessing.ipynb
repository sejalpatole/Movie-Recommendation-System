```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies.columns
movies = movies[['id', 'title', 'genres', 'keywords', 'overview', 'production_companies', 'tagline']]
```

```python
credits = credits[['movie_id', 'cast', 'crew']]
```

```python
df = pd.merge(movies, credits, left_on = 'id', right_on = 'movie_id', how='inner')
```

```python
df.shape
```

```python
df.drop(columns = ['movie_id'], inplace = True)
```

```python
df.to_csv("data.csv", index = False)
```
